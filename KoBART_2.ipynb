```python
import torch
from transformers import BartForConditionalGeneration, PreTrainedTokenizerFast
```

```python
model_path = "/content/drive/MyDrive/SKT_KoBART/kobart_summary_model_epoch_1"
model = BartForConditionalGeneration.from_pretrained(model_path)
tokenizer = PreTrainedTokenizerFast.from_pretrained('gogamza/kobart-base-v1')
```

```python
model.eval()
```

```python
text = """[   ] ' 5G '    "",   ' ' 5G"" ""    ""  .   8    5G   ""5G   ,      "" ""5G   "" .      ''  , 5G 4          .   ""5G   ,   , (), ,  TV(CCTV)           "" .     5G         .  2026  5G   1161    .     529   2      .       5G       .   ""'  , 5G  ?'   "" ""4  ''      "" .  ""         ,        "" .        5G      . 5G             .   "" 360 ,  427              "" ."""

#   
inputs = tokenizer([text], max_length=1024, return_tensors='pt', truncation=True)

#   
summary_ids = model.generate(inputs['input_ids'], num_beams=4, max_length=150, early_stopping=True)

#   
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

print("Original Text: ", text)
print("Summary: ", summary)
```

```python
from rouge_score import rouge_scorer

def calculate_rouge_scores(reference, hypothesis):
    # ROUGE Scorer 
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

    # ROUGE  
    scores = scorer.score(reference, hypothesis)

    return scores

#      
reference_summary = """8   5G     5G    ",   ' '"  5G                ."""
generated_summary = """  8    5G   "5G   ,      " "5G   " ."""

# ROUGE  
rouge_scores = calculate_rouge_scores(reference_summary, generated_summary)

# ROUGE  
print(f"ROUGE-1: {rouge_scores['rouge1'].fmeasure}")
print(f"ROUGE-2: {rouge_scores['rouge2'].fmeasure}")
print(f"ROUGE-L: {rouge_scores['rougeL'].fmeasure}")
```

```python
from nltk.translate.bleu_score import sentence_bleu

reference_summary = """8   5G     5G    ",   ' '"  5G                ."""
generated_summary = """  8    5G   "5G   ,      " "5G   " ."""

reference = [reference_summary.split()]  #     .
hypothesis = generated_summary.split()  #     .

bleu_score = sentence_bleu(reference, hypothesis)

print(f"BLEU Score: {bleu_score}")
```

```python
import pandas as pd

# CSV  
file_path = "/content/sample_data/news_contents_0820_0822.csv"
df = pd.read_csv(file_path)
```

```python
from transformers import BartForConditionalGeneration, PreTrainedTokenizerFast

#   
model_path = "/content/drive/MyDrive/SKT_KoBART/kobart_summary_model_epoch_1"
model = BartForConditionalGeneration.from_pretrained(model_path)
tokenizer = PreTrainedTokenizerFast.from_pretrained('gogamza/kobart-base-v1')

def summarize_article(content):
    inputs = tokenizer.encode(content, return_tensors='pt', max_length=1024, truncation=True)
    summary_ids = model.generate(inputs, max_length=150, min_length=30, length_penalty=2.0, num_beams=4, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

#    
df['summary'] = df['content'].apply(summarize_article)
```

```python
output_file_path = "/content/news_contents_0820_0822.csv"
df.to_csv(output_file_path, index=False)
```

```python
import pandas as pd

# CSV  
contents_csv_path = '/content/news_contents_0820_0822 (1).csv'
titles_csv_path = '/content/2024_0820_0822_.csv'

#  
contents_df = pd.read_csv(contents_csv_path)
titles_df = pd.read_csv(titles_csv_path)

#   
contents_df = contents_df[['url', 'content', 'summary']]
titles_df = titles_df[['Title', 'Original link', 'PublicationDate']]

# URL    
merged_df = pd.merge(contents_df, titles_df, left_on='url', right_on='Original link', how='inner')

#        
merged_df = merged_df.rename(columns={'Title': 'title', 'PublicationDate': 'article_date'})
merged_df = merged_df[['title', 'content', 'summary', 'article_date']]

#    (YYYY-MM-DD)
merged_df['article_date'] = pd.to_datetime(merged_df['article_date']).dt.strftime('%Y-%m-%d')

#   CSV  
final_csv_path = '/content/merged_news_articles.csv'
merged_df.to_csv(final_csv_path, index=False)

print(" CSV  :", final_csv_path)
```
